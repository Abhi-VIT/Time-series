# **Random Walk**

**‚óè Create and plot a random walk of 1000 timesteps. Set appropriate seed. (Read on
it!)**

```python
np.random.seed(42)
rw = np.random.choice([-1,1,0],1000)
w = rw
rw
```

```python
for i in range(2, len(rw)):
    rw[i] = rw[i-1] + w[i] + 0
```

```python
data_rw = pd.Series({'Walk' : rw})
data_rw
```

**‚óè create a line plot of the given series.**

```python
type(data_rw)
```

```python
data_rw.describe()
```

```python
plt.figure(figsize =(40,24))
sns.scatterplot(data_rw.Walk)
plt.plot(data_rw.Walk)
plt.show()
```

By ploiting the graph without dift 

**‚óè Create the ACF and PACF plots.**

```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
plot_acf(data_rw.Walk, lags = len(data_rw.Walk)/2)
plt.show()
```

Autocorrelation decreases and then goes into the negative 

```python
plot_pacf(data_rw.Walk, lags = len(data_rw.Walk)/2)
plt.show()
```

Partial Auto correlation sudden decrease which show the randomness of the data.

**‚óè Print the summary statistics**

```python
mean_data = data_rw.Walk.mean().round(2)
print(f'lag t : {mean_data}')
l = [10, 50, 100, 250, 300, 350 , 400, 500, 600, 700, 800, 999]
for lag in l:
    mean_data = data_rw.Walk[lag:].mean().round(2)
    print(f'mean of data at lag {lag} : {mean_data}')
```

```python
var_data = data_rw.Walk.var().round()
print(f'lag t : {var_data}')
l = [10, 50, 100, 250, 300, 350 , 400, 500, 600, 700, 800, 999]
for lag in l:
    var_data = data_rw.Walk[lag:].var().round()
    print(f'varience of data at lag {lag} : {var_data}')
```

# **Key Differences Between White Noise and Random Walk**
### Plots:
#### **White Noise:** Appears as scattered points without a clear pattern.
#### **Random Walk:** Shows a cumulative trend, leading to a more structured, drifting appearance.
### **Statistical Properties:**

#### **White Noise:**
#### **Mean:** Constant around 0
#### **Variance:** Constant over time (stationary)
### No correlation between time steps
#### **Random Walk:**
#### **Mean:** Varies over time (non-stationary)
#### **Variance:** decrease with time
#### **Strong correlation between successive values**
